```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# concatenates all years files into TradeData_all.csv

directory = './data/UNComtrade/'
output_file = os.path.join(directory, 'TradeData_all.csv')
all_dfs = []

for filename in os.listdir(directory):
    if filename.endswith('.csv') and filename != 'TradeData_all.csv':
        file_path = os.path.join(directory, filename)

        try:
            df = pd.read_csv(file_path, encoding='Windows-1252')
        except UnicodeDecodeError:
            df = pd.read_csv(file_path, encoding='latin1')
        
        all_dfs.append(df)
        print(f"Loaded {filename} with {len(df)} rows.")

# Concatenate all dataframes and save to CSV
if all_dfs:
    combined_df = pd.concat(all_dfs, ignore_index=True)
    combined_df.to_csv(output_file, index=False)
    print(f"Successfully combined {len(all_dfs)} files into {output_file}")
```
